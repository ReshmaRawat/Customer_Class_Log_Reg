```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# from sklearn.preprocessing import ProfileReport

from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, f1_score,confusion_matrix,classification_report
```

```python
train_data = pd.read_csv('existing_base_train.csv')
```

```python
train_data.head()
```

```python
train_data.info()
```

```python
train_data.isnull().sum()
```

```python
for col in train_data.columns:
    if train_data[col].dtype=='object':
        print(col)
        print(train_data[col].value_counts())
```

```python
train_data['children'] = train_data['children'].replace('Zero' , int(0))
train_data['children'] = train_data['children'].replace('4+' , int(4))
```

```python
train_data['children'] = pd.to_numeric(train_data['children'])
```

```python
train_data.info()
```

```python
train_data.corr()
```

```python
train_data[['Home_Loan' , 'Online_Purchase_Amount']].corr()
```

```python
le = LabelEncoder()
```

```python
x = train_data.drop(['Revenue_Grid' , 'REF_NO'],axis = 1)
y = train_data['Revenue_Grid']
```

```python
for col in x.columns:
    if x[col].dtype == 'object':
        x[col] = le.fit_transform(x[col])
```

```python
x.head()
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(x,y)
```

```python
y_hat = log_reg.predict(x)
```

```python
y_hat_train = pd.DataFrame(y_hat)
```

```python
report = classification_report(y, y_hat_train)
```

```python
print(report)
```
